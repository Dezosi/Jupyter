# Brilliants

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

data = pd.read_csv("P1_diamonds.csv", header = 0)
data = data.drop("Unnamed: 0", axis = 1)
data.head()
```

# Графики

```python
plt.hist(data["price"], bins = 150)

plt.title("Частотный график: цены на бриллианты\n", fontdict={'fontsize':20, 'fontweight':20, 'color': 'Green'})
plt.xlabel("Цена", fontdict={'fontsize':12, 'fontweight':20, 'color': 'Brown'})
plt.ylabel("Количество", fontdict={'fontsize':12, 'fontweight':20, 'color': 'Brown'})

plt.show()
```

```python
# Построение гистограммы
sns.histplot(data=data, x="price", bins=150, color="green")

# Настройка заголовка и подписей осей
plt.title("Частотный график: цены на бриллианты\n", fontsize=20, fontweight='bold', color='Green')
plt.xlabel("Цена", fontsize=12, fontweight='bold', color='Brown')
plt.ylabel("Количество", fontsize=12, fontweight='bold', color='Brown')

# Показ графика
plt.show()
```

```python
groupdata = data.groupby("cut").count()
groupdata

x_vals = []
y_vals = []
for i in [0, 1, 4, 3, 2]:
    x_vals.append(groupdata.index[i])
    y_vals.append(groupdata.iloc[i,0])

plt.bar(x_vals, y_vals)
plt.title("Частотный график: огранка бриллиантов\n", fontdict={'fontsize':20, 'fontweight':20, 'color': 'Green'})
plt.xlabel("Огранка", fontdict={'fontsize':12, 'fontweight':20, 'color': 'Brown'})
plt.ylabel("Количество", fontdict={'fontsize':12, 'fontweight':20, 'color': 'Brown'})

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

groupdata = data.groupby("cut").count().iloc[[0, 1, 4, 3, 2]]
x_vals = groupdata.index
y_vals = groupdata.iloc[:, 0]

sns.barplot(x=x_vals, y=y_vals, palette="husl", dodge=False, legend=False)

plt.title("Частотный график: огранка бриллиантов\n", fontsize=20, fontweight='bold', color='Green')
plt.xlabel("Огранка", fontsize=12, fontweight='bold', color='Brown')
plt.ylabel("Количество", fontsize=12, fontweight='bold', color='Brown')

plt.show()
```
